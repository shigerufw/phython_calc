## D- Calculate the margin of sales and rate it 

```python
vendas =  [200 ,500 ,300 ,1000.0 ,250 ,980 ,321 ,267]
custo_RS = [ 121.0,347 ,235 ,890 ,206 ,321 ,130 ,230]
df_d = pd.DataFrame({'Vendas (R$)': vendas, 'Custo': custo_RS})
df_d
```

#### 4.03 - calculating "Margem %"

```python
df_d['Margem %']= (df_d["Vendas (R$)"]-df_d['Custo'])/df_d["Vendas (R$)"]
df_d
```

#### 4.04 - rating the "Margem"

```python
def return_rate(margem):
    """
    input = margem (number)
    return = classification ["Ruim","Boa","Excelente"]
    """
    if margem < 0.2:
        return 'Ruim'
    elif margem < 0.35:
        return "Boa"
    else:
        return "Excelente"
```

```python
test = [0, .24,.40, -1]
list(map(return_rate, test))
```

```python
df_d['Classificação'] = df_d["Margem %"].apply(return_rate)
df_d
```

##### 5.02.03 - PCT do valor de Financiamento 

```python
pd_01['pct_fin'] = (pd_01['Valor total Financiamento'])/pd_01['Valor total']
```

```python
plt.figure(figsize = (12,8))
plt.title("PCT Financiamento por classificação")
sns.distplot(pd_01[pd_01['Classificação'] == 'A']['pct_fin'],kde = 0, label = 'A')
sns.distplot(pd_01[pd_01['Classificação'] == 'B']['pct_fin'],kde = 0, label = 'B')
sns.distplot(pd_01[pd_01['Classificação'] == 'C']['pct_fin'],kde = 0, label = 'C')
sns.distplot(pd_01[pd_01['Classificação'] == 'D']['pct_fin'],kde = 0, label = 'D')
sns.distplot(pd_01[pd_01['Classificação'] == 'E']['pct_fin'],kde = 0, label = 'E')
plt.legend()
```

```python
m_f_a = pd_01[pd_01['Classificação'] == 'A']['pct_fin'].mean()
m_f_b = pd_01[pd_01['Classificação'] == 'B']['pct_fin'].mean()
m_f_c = pd_01[pd_01['Classificação'] == 'C']['pct_fin'].mean()
m_f_d = pd_01[pd_01['Classificação'] == 'D']['pct_fin'].mean()
m_f_e = pd_01[pd_01['Classificação'] == 'E']['pct_fin'].mean()

med_f_a = pd_01[pd_01['Classificação'] == 'A']['pct_fin'].median()
med_f_b = pd_01[pd_01['Classificação'] == 'B']['pct_fin'].median()
med_f_c = pd_01[pd_01['Classificação'] == 'C']['pct_fin'].median()
med_f_d = pd_01[pd_01['Classificação'] == 'D']['pct_fin'].median()
med_f_e = pd_01[pd_01['Classificação'] == 'E']['pct_fin'].median()



rank = [(m_f_a,'a',med_f_a),
        (m_f_b,'b',med_f_b),
        (m_f_c,'c',med_f_c),
        (m_f_d,'d',med_f_d),
        (m_f_e,'e',med_f_e)]
rank.sort()
for i in range(len(rank)):
    print('Pct financeira média %s = %.2f - pct financeira mediana  = %.2f  ' %(rank[i][1],rank[i][0]*100, rank[i][2]*100))
```

<b> B apresenta menor valor percentual médio para financiamento, C apresenta maior valor percentual médio para financiamento anual </b>

```python
plt.figure(figsize = (12,8))
plt.title("PCT Financiamento separado por grupo")
sns.distplot(pd_01[pd_01['Grupo de Produto'] == 1]['pct_fin'],kde = 0, label = '1')
sns.distplot(pd_01[pd_01['Grupo de Produto'] == 3]['pct_fin'],kde = 0, label = '3')
plt.legend()
```
